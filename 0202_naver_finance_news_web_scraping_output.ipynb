## 위의 과정을 함수로 만듭니다.

````python
# get_url item_code, page_no 를 넘기면 url 을 반환하는 함수
def get_url(item_code, page_no):
    """
    item_code, page_no 를 넘기면 url 을 반환하는 함수
    """
    # url
    url = f"https://finance.naver.com/item/news_news.nhn?code={item_code}&page={page_no}&sm=title_entity_id.basic&clusterId="
    return url
````

## 뉴스 한 페이지를 수집하는 함수
```
1) URL 을 받아옴
2) read_html 로 테이블 정보를 받아옴
3) 데이터프레임 컬럼명을 ["제목", "정보제공", "날짜"]로 변경
4) temp_list 에 데이터프레임을 추가
5) concat 으로 리스트 병합하여 하나의 데이터프레임으로 만들기
6) 결측치 제거
7) 연관기사 제거
8) 데이터프레임 반환
```

````python
# get_one_page_news 함수 만들기
def get_one_page_news(item_code, page_no):
    """
    get_url 에 item_code, page_no 를 넘겨 url 을 받아오고
    뉴스 한 페이지를 수집하는 함수
    1) URL 을 받아옴
    2) read_html 로 테이블 정보를 받아옴
    3) 데이터프레임 컬럼명을 ["제목", "정보제공", "날짜"]로 변경
    4) temp_list 에 데이터프레임을 추가
    5) concat 으로 리스트 병합하여 하나의 데이터프레임으로 만들기
    6) 결측치 제거
    7) 연관기사 제거
    8) 데이터프레임 반환
    """
    url = get_url(item_code, page_no)
    table = pd.read_html(url, encoding="cp949")
    
    temp_list = []
    cols = table[0].columns

    for news in table[:-1]:
        news.columns = cols
        temp_list.append(news)
        
    df_temp = pd.concat(temp_list)
    df_temp = df_temp.dropna(how="all", axis=0).dropna(how="all", axis=1)
    df = df_temp[~df_temp["제목"].str.contains("연관기사")]
    
    return df
````

* 아래 코드를 실행했을 때 수집된 데이터프레임이 출력된다면 잘 만들어진 것입니다.

````python
page_no = 3

temp = get_one_page_news(item_code, page_no)
temp
````

## 반복문을 사용해 10페이지까지 수집
* 위에서 만든 함수를 활용합니다.
* time.sleep() 을 통해 데이터를 쉬었다 가져옵니다.

````python
import time
from tqdm import trange

page_no = 0
news_list = []

for page_no in trange(1, 11):
    df_temp = get_one_page_news(item_code, page_no)
    time.sleep(0.01)
    news_list.append(df_temp)
````

## 위에서 수집한 데이터는 news_list 라는 변수에 들어있습니다. 판다스를 활용해 하나의 데이터프레임으로 만들어 주세요.

    * axis=1 컬럼을 기준으로 인덱스가 같은 값을 옆으로 붙여 새로운 컬럼을 만듦

````python
# 1) 데이터 프레임 하나로 합치고 인덱스 번호가 제각각인것을 볼 수 있습니다.
# 2) reset_index() 를 하면 기존 인덱스가 index 컬럼으로 생기고 새로운 인덱스가 생깁니다.
# 3) reset_index(drop=True) 기존 인덱스를 제거하고 새로 생성한 인덱스만 남깁니다.
df_news = pd.concat(news_list).reset_index(drop=True)
````

````python
import pandas as pd
import time
from tqdm import trange


class Krx:
    def __init__(self):
        self.__url = "http://kind.krx.co.kr/corpgeneral/corpList.do?method=download&searchType=13"

    def find_all(self):
        # krx에서 모든 상장 회사들의 종목 코드를 가져오기.
        df_stock = pd.read_html(self.__url, header=0)[0][["회사명", "종목코드"]]
        my_columns = {"회사명": "Name", "종목코드": "Symbol"}

        df_stock = df_stock.rename(columns=my_columns)
        df_stock["Symbol"] = [f"{code:06d}" for code in df_stock["Symbol"]]

        return df_stock

    def find(self, company_name):
        # 가져온 상장회사 목록에서 원하는 회사의 종목코드 가져오기.
        df_stock = self.find_all()
        idx = df_stock["Name"] == company_name
        symbol = df_stock[idx]["Symbol"]

        if symbol.empty:
            raise ValueError("종목 코드를 가져오는데 실패했습니다. (올바른 회사명인지 확인해주세요.)")

        return symbol.values[0]


class NaverFinancialNews:
    def __init__(self, company_symbol):
        self.company_symbol = company_symbol

    def __repr__(self):
        return f"NaverFinancialNews(company_symbol={self.company_symbol!r})"

    def get_one_page(self, page_no):
        # 지정한 하나의 페이지 뉴스만  가져오기
        url = f"https://finance.naver.com/item/news_news.nhn?code={self.company_symbol}&page={page_no}&sm=title_entity_id.basic&clusterId="

        df = pd.read_html(url, encoding="cp949")[0]

        news = df[df["제목"].str.contains("연관기사") == False]

        return news

    def get_multiple_pages(self, page_no):
        # 여러 페이지의 뉴스 가져오기.
        news = pd.DataFrame()

        for page in trange(1, page_no + 1):
            time.sleep(0.1)
            news = news.append(self.get_one_page(page))

        return news


def naver_financial_news(company_name):
    krx = Krx()
    code = krx.find(company_name)
    news = NaverFinancialNews(code)
    return news


if __name__ == "__main__":
    news = naver_financial_news("현대자동차")
    print(news.get_one_page(6))
    print(news.get_multiple_pages(4))
````
